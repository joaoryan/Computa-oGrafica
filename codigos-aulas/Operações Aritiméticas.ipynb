1- Aplique HDR a alguma imagem de https://hdr-photographer.com/hdr-photos-to-play-with/.

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt


sunset_1 = np.array(Image.open('IMG_4356.jpg'))
sunset_2 = np.array(Image.open('IMG_4357.jpg'))
sunset_3 = np.array(Image.open('IMG_4358.jpg'))
sunset_4 = np.array(Image.open('IMG_4359.jpg'))

def hdr(imagem1,imagem2,imagem3,imagem4):
    
    sunset_avg = ( imagem1.astype(np.float64)
                + imagem2.astype(np.float64)
                + imagem3.astype(np.float64)
                + imagem4.astype(np.float64)) / 4
    sunset_avg = sunset_avg.astype(np.uint8)
    
    return sunset_avg
        

sunset_avg = hdr(sunset_1,sunset_2,sunset_3,sunset_4)
#Image.fromarray(spunset_avg).save('sunset_avg.png')

plt.figure(figsize=(16, 16))
plt.subplot(3, 2, 1)
plt.imshow(sunset_1)
plt.subplot(3, 2, 2)
plt.imshow(sunset_2)
plt.subplot(3, 2, 3)
plt.imshow(sunset_3)
plt.subplot(3, 2, 4)
plt.imshow(sunset_4)
plt.subplot(3, 2, 5)
plt.imshow(sunset_avg)
```

2- Crie uma função blend, que recebe como parâmetros img1, img2 e c, que realiza uma operação de blending entre duas imagens e retorna a imagem resultante. Em seguida, busque duas imagens de mesmo tamanho e realize um blending, com coeficiente de sua escolha.

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

img1 = np.array(Image.open('ex_blend1.jpg'))
img2 = np.array(Image.open('ex_blend2.jpg'))

C = 0.4

def blending(imagem_1,imagem_2,c_blending):
    
    imagem_blending = (c_blending * img1) + ((1 - c_blending) * img2)
    imagem_blending = imagem_blending.astype(np.uint8)
    
    return imagem_blending

plt.figure(figsize=(16, 16))
imagem_blending = blending(img1,img2,C)
plt.subplot(2, 2, 1)
plt.imshow(img1)
plt.subplot(2, 2, 2)
plt.imshow(img2)
plt.subplot(2, 2, 3)
plt.imshow(imagem_blending)
```

3- Aumente (ou diminua, caso prefira) a iluminação da imagem gerada acima. Salve o resultado no disco.

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

img = np.array(Image.open('ex_blend1.jpg'))

def realce_de_cor(img_brilho,teste_brilho):
    
    if(teste_brilho == True):
        
        photo_mul = np.clip(img_brilho * 1.5, 0, 255).astype(np.uint8)
        
        return photo_mul    
    
    else:
        
        photo_div = (img_brilho / 2.0).astype(np.uint8)
    
        return photo_div

photo_mul = realce_de_cor(img,True)    
photo_div = realce_de_cor(img,False)
plt.figure(figsize=(16, 16))
#(linha,coluna,índice)
plt.subplot(1, 2, 1)
plt.imshow(photo_mul)
plt.subplot(1, 2, 2)
plt.imshow(photo_div)
Image.fromarray(photo_mul).save('com_brilho.jpg')
```
