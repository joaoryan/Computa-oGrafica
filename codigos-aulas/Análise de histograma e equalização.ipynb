Escolha uma imagem colorida na internet e exiba os seguintes histogramas, utilizando plt.subplot(2, 1, x):

Em escala de cinza utilizando um dos tr√™s m√©todos vistos em aulas anteriores (em x = 2).
Lembrando:

1 - Lightness: (ùëöùëéùë•(ùëü,ùëî,ùëè)+ùëöùëñùëõ(ùëü,ùëî,ùëè)) / 2

2 - Average: (ùëü+ùëî+ùëè) / 3

 
3 - Luminosity: 0.2126 √ó ùëü + 0.7152 √ó ùëî + 0.0722 √ó ùëè

```python
from PIL import Image 
import matplotlib.pyplot as plt
import numpy as np

img = np.array(Image.open('girl.jpg'))

#1¬∫)Imagem em RGB
def histogram_gray(img):
    (l, c) = img.shape
    # Histograma com 1 linha e 256 colunas, preenchido com 0.
    histogram = np.zeros(256, dtype=np.int32)

    for i in range(l):
        for j in range(c):
            # Normalizar intensidade do pixel 
            intensity = img[i, j]
            # Incrementar de 1 a quantidade de pixels no histograma,
            # utilizando a intensidade como √≠ndice.
            histogram[intensity] += 1
            
    return histogram

def show_histogram(img):
    histogram_r = histogram_gray(img[:, :, 0])
    histogram_g = histogram_gray(img[:, :, 1])
    histogram_b = histogram_gray(img[:, :, 2])
    
    plt.figure(figsize=(16, 16))
    plt.subplot(3, 2, 1)
    plt.imshow(img, cmap='gray', vmin=0, vmax=255)
    plt.subplot(3, 2, 2)
    plt.bar(range(256), histogram_r, color='red')
    plt.bar(range(256), histogram_g, color='green')
    plt.bar(range(256), histogram_b, color='blue')
    
#2¬∫)Mostando o histograma da imagem em grayscale
def grayscale(img_np):
    (l, c, p) = img_np.shape

    img_avg = np.zeros(shape=(l, c), dtype=np.uint8)
    for i in range(l):
        for j in range(c):
            r = float(img_np[i, j, 0])
            g = float(img_np[i, j, 1])
            b = float(img_np[i, j, 2])
        
            img_avg[i, j] = (r + g + b) / 3
     
    return img_avg

def histogram_gray(img):
    (l, c) = img.shape
    # Histograma com 1 linha e 256 colunas, preenchido com 0.
    histogram = np.zeros(256, dtype=np.int32)

    for i in range(l):
        for j in range(c):
            # Normalizar intensidade do pixel 
            intensity = img[i, j]
            # Incrementar de 1 a quantidade de pixels no histograma,
            # utilizando a intensidade como √≠ndice.
            histogram[intensity] += 1
            
    return histogram         
def show_histogram_gray(img):
    histogram = histogram_gray(img)
    plt.figure(figsize=(16, 16))
    plt.subplot(3, 2, 3)
    # vmin e vmax n√£o permitem que o matplotlib fa√ßa normaliza√ß√£o das cores.
    plt.imshow(img, cmap='gray', vmin=0, vmax=255)
    plt.subplot(3, 2, 4)
    plt.bar(range(256), histogram)
    
show_histogram(img)
img_avg = grayscale(img)
show_histogram_gray(img_avg)
```
