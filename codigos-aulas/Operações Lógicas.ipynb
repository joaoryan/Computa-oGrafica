Operação NOT

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

scene1 = np.array(Image.open("scene1.gif"))
scene2 = np.array(Image.open("scene2.gif"))

# Mais uma forma de aplicar threshold:
def threshold(img, thresh):
    out = img.copy()
    out[img >  thresh] = 255
    out[img <= thresh] = 0
    return out

bin1 = threshold(scene1, 63)
bin2 = threshold(scene2, 63)

def logica_not(bin1,bin2):
    
    bin1_not = np.bitwise_not(bin1) #bitwise- bit a bit
    bin2_not = np.invert(bin2)
    
    return bin1_not,bin2_not
    
bin1_not, bin2_not = logica_not(bin1,bin2)
plt.figure(figsize=(16, 16))
plt.subplot(2, 2, 1)
plt.imshow(bin1, cmap='gray')
plt.subplot(2, 2, 2)
plt.imshow(bin2, cmap='gray')
plt.subplot(2, 2, 3)
plt.imshow(bin1_not, cmap='gray')
plt.subplot(2, 2, 4)
plt.imshow(bin2_not, cmap='gray')
```

Operação AND

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

scene1 = np.array(Image.open("scene1.gif"))
scene2 = np.array(Image.open("scene2.gif"))

# Mais uma forma de aplicar threshold:
def threshold(img, thresh):
    out = img.copy()
    out[img >  thresh] = 255
    out[img <= thresh] = 0
    return out

bin1 = threshold(scene1, 63)
bin2 = threshold(scene2, 63)


def logica_and(bin1,bin2):
    #0 - preto, 1 - branco, realiza a operação lógica
    bin_and = np.bitwise_and(bin1, bin2)
    return bin_and
   
    
bin_and = logica_and(bin1,bin2)
plt.figure(figsize=(16, 16))
plt.subplot(2, 2, 1)
plt.imshow(bin1, cmap='gray')
plt.subplot(2, 2, 2)
plt.imshow(bin2, cmap='gray')
plt.subplot(2, 2, 3)
plt.imshow(bin_and, cmap='gray')
```
