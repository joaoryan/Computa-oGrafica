# Operações lógicas em imagens
## Aula 8

* Operações lógicas **E** (AND), **OU** (OR), **NÃO** (NOT) e **XOU** (XOR);
* E, OU e NÃO também são frequentemente chamadas de **interseção**, **união** e **complemento**, respectivamente;
* Normalmente são utilizadas **imagens binárias** para estas operações;
* Operam bit a bit em cada valor dentro do array.

## Exercícios

1. Escolha algumas imagens e realize algumas das operações desta aula nelas.

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

img = np.array(Image.open("img.jpeg"))[:,:,:3]
l,c,p = img.shape
mask = np.zeros(shape = (l,c,p), dtype = np.uint8)

mask[img.shape[0] // 4:img.shape[0] * 3 // 4,
     img.shape[1] // 4:img.shape[1] * 3 // 4,
     :] = 255


newImg = mask & img
plt.figure(figsize=(16, 16))
plt.subplot(2, 2, 1)
plt.imshow(img)
plt.subplot(2, 2, 2)
plt.imshow(mask)
plt.subplot(2, 2, 3)
plt.imshow(newImg)
```

2. **(Opcional)** Realize uma operação de threshold nas imagens e repita as operações acima.

```python
def threshold(img, thresh):
    out = img.copy()
    out[img >  thresh] = 255
    out[img <= thresh] = 0
    return out

import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

img = np.array(Image.open("img.jpeg"))[:,:,:3]
l,c,p = img.shape
mask = np.zeros(shape = (l,c,p), dtype = np.uint8)

mask[img.shape[0] // 4:img.shape[0] * 3 // 4,
     img.shape[1] // 4:img.shape[1] * 3 // 4,
     :] = 255


newImg = mask & img
plt.figure(figsize=(16, 16))
plt.subplot(2, 2, 1)
plt.imshow(img)
plt.subplot(2, 2, 2)
plt.imshow(mask)
plt.subplot(2, 2, 3)
plt.imshow(newImg)
plt.subplot(2, 2, 4)
bin = threshold(newImg, 63)
plt.imshow(bin, cmap='gray')
```
