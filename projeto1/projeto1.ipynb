### OpenCV (cv2)
- **Introdução da técnica escolhida**: OpenCV é uma biblioteca de visão computacional de código aberto.
- **Por quem foi criada**: Foi criada por Gary Bradsky com o apoio da Intel em 2000.
- **Quando**: Foi lançada oficialmente em 2000.
- **Por que**: Foi criada para fornecer uma infraestrutura comum para aplicações de visão computacional e para acelerar o uso da percepção de máquina na comercialização de produtos.
- **Como funciona**: OpenCV possui módulos para processamento de imagem, estruturas de dados, álgebra linear, interfaces gráficas básicas e muito mais.
- **Pontos fortes**: É eficiente, possui algoritmos de última geração e é amplamente utilizada na comunidade de visão computacional.
- **Pontos fracos**: Pode ser complexa para iniciantes e a documentação pode ser difícil de entender.


## Codigo

1. `import cv2`: Este comando importa a biblioteca OpenCV (Open Source Computer Vision Library), que é uma biblioteca de programação voltada para o processamento de imagens e visão computacional.

2. `import matplotlib.pyplot as plt`: Este comando importa a biblioteca Matplotlib, que é uma biblioteca para a criação de visualizações estáticas, animadas e interativas em Python. Neste caso, estamos usando o módulo `pyplot` para exibir as imagens.

3. `img = cv2.imread('imagem1.png')`: Este comando usa a função `imread` do OpenCV para ler a imagem 'imagem1.png' e armazená-la na variável `img` como uma matriz NumPy.

4. `img_suavizada = cv2.GaussianBlur(img, (5, 5), 0)`: Este comando usa a função `GaussianBlur` do OpenCV para aplicar um filtro de borramento Gaussiano à imagem. O segundo argumento `(5, 5)` especifica o tamanho do kernel (a "janela" que o filtro usa ao redor de cada pixel). O terceiro argumento `0` especifica o desvio padrão no eixo X e no eixo Y; quando é definido como `0`, ele é calculado com base no tamanho do kernel.

5. `cv2.imwrite('imagem_suavizada.png', img_suavizada)`: Este comando usa a função `imwrite` do OpenCV para escrever a imagem suavizada em um arquivo chamado 'imagem_suavizada.png'.

6. `img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)` e `img_suavizada = cv2.cvtColor(img_suavizada, cv2.COLOR_BGR2RGB)`: Estes comandos usam a função `cvtColor` do OpenCV para converter as imagens do espaço de cores BGR (Blue, Green, Red), que é o padrão do OpenCV, para o espaço de cores RGB (Red, Green, Blue), que é o padrão do Matplotlib.

7. As linhas restantes com `plt.figure`, `plt.subplot` e `plt.imshow` usam funções do Matplotlib para criar uma figura, adicionar subplots à figura e exibir as imagens nos subplots.


```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('imagem1.png')

img_suavizada = cv2.GaussianBlur(img, (5, 5), 0)

cv2.imwrite('imagem_suavizada.png', img_suavizada)

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_suavizada = cv2.cvtColor(img_suavizada, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(16, 16))
plt.subplot(2, 2, 1)
plt.imshow(img)
plt.subplot(2, 2, 2)
plt.imshow(img_suavizada)
```
